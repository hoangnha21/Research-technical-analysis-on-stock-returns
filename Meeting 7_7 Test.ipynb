```python
# Load the dataset from Google Drive
import pandas as pd

dataset_path = '/content/drive/MyDrive/2023SummerResearch/SPY.csv'  # Update the path according to your dataset location
SP500data = pd.read_csv(dataset_path)
```

```python
SP500data
```

```python
#Calculate the 60-period moving average using the 'Close' price column:
lookback = 60
moving_average = SP500data['Close'].rolling(lookback).mean()
```

```python
#Calculate the price deviations around the moving average:
deviations = SP500data['Close'] - moving_average
```

```python
#Count the number of up periods relative to the total number of periods:
up_periods = np.sum(deviations > 0)
total_periods = len(deviations)
```

```python
#Calculate the Trend Intensity Indicator:
trend_intensity_indicator = up_periods / total_periods
```

```python
print("Trend Intensity Indicator:", trend_intensity_indicator)
```
